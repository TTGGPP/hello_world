```python
def get_data_from_db(query):
    conn = cx_Oracle.connect('hoseo', 'hoseo', 'localhost:1521/xe')
    df = pd.read_sql(query, conn)
    conn.close()
    return df
```

```python
# 지역별 전체 매매건수
apt_price = """
            SELECT REGION, COUNT(*)
            FROM APT_DEAL_PRICE
            GROUP BY REGION
            """
```

```python
apt = get_data_from_db(apt_price)
apt.info()
apt.head()
```

```python
# 지역별 사이즈별 매매건수
apt_price2 = """
             SELECT REGION, COUNT(*)
             FROM APT_DEAL_PRICE
             WHERE APT_SIZE = 30 
             GROUP BY REGION
             """
```

```python
apt2 = get_data_from_db(apt_price2)
apt2.info()
apt2.head()
```

```python
# 지역별 평균 거래금액
apt_price3 = """
             SELECT REGION, AVG(CONTRACT_PRICE)
             FROM APT_DEAL_PRICE
             GROUP BY REGION
             """
```

```python
# 세부 지역별 평균 거래금액
apt_price4 = """
             SELECT REGION_SUB, AVG(APT_SIZE)
             FROM APT_DEAL_PRICE
             GROUP BY REGION_SUB
             """
```
